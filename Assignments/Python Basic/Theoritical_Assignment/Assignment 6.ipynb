**1. What are escape characters?**

Ans. In python there are some strings or character which are illegal. We are not able to use them So in order to use those strings or character we need the escape character. This character is basically backslash '\' followed by the character which you want to have in your syntax.

Eg: print(' I don\'t want to use any other /tlanguage')

o/p: 'I don't want to use any other   language')

Other: 
\' -->Single Quote	
\\ -->Backslash	
\n -->New Line	
\r -->Carriage Return	
\t --> Tab	
\b -->Backspace
